## 1-2:機械学習の位置づけと流れ

### 機械学習とは
大量のデータから、機械学習アルゴリズムによってデータの特性を見つけて、予測などを行う計算式の塊（＝モデル）を作る。
このモデルから、入力したデータ以外の未知のデータで数値予測や、入力データのカテゴライズなんかができる。

### 機械学習以外の選択肢
* ルールベース
if文で条件分岐するイメージ。パラメータの数を増やすと記述が難しくなる。
* 統計的手法
データから統計的な数値を求め、それらを元に予測するアプローチ。機械学習のアルゴリズムと親和性が高い。

### 機械学習のタスク

* 教師あり学習

    * 正解のラベルデータが存在する時に用いる
    * 目的変数と説明変数
    * 回帰と分類に分かれる。
        * 回帰：正解ラベルが連続値、結果も連続値
        * 売上金額の予測とか気温の予測とか
        * 分類：正解ラベルが連続値ではない
        
* 教師なし学習

    * クラスタリング
        * 与えられたデータの中からグルーピング
    * 次元圧縮
        * 説明変数の次元数を減らす

* 強化学習

    * 将棋とか囲碁とかのAI。今回は割愛。

### 機械学習の処理の手順（教師あり分類の場合）

* データ入手
データを手に入れて、目的の機械学習に使えることを確認する(NumPy,Pandas)

* データ加工
初期加工。データ型整えたり、データの連結とか、欠損値の対応とか。(NumPy,Pandas)

* データ可視化
まとまったデータを、表やグラフで可視化。統計量の確認。(Matplotlib,Pandas)

* アルゴリズム選択
アルゴリズムを選択する。機械学習で最も難しい作業の一つ。scikit-learnが多く活用される。

## 1-3:データ分析に使う主なパッケージ

### パッケージとは？
Pythonに機能を追加したり支援するためのもの

### サードパーティ製パッケージ

* **Jupyter Notebook**
今のこれ。ipynb形式のファイル。中身はJSONらしい。

* **NumPy**
数値計算を行う。配列とか行列とかが得意。
中身はC言語で実装されてるらしい。高速。
配列内部のデータに対して直接データ変換ができたり、
行列とベクトルの足し算を、分解せずに高速で行える。速い。

* **pandas**
NumPyを基礎とした、DataFrame構造を提供する。
表形式の2次元データを柔軟に取り扱えるのが特徴。
データの変換とか加工とか、データ分析でよく使う作業が効率的に処理できる。

* **Matplotlib**
データの可視化を行う。
折れ線グラフとか棒グラフとかヒストグラムとか。見てわかるようになる。

* **scikit-learn**
機械学習のアルゴリズムとか評価ツールとかが一緒になったセット。
使いやすいらしい。
アルゴリズムは実装して、評価ツールだけ使う、みたいな活用もあり。

* **SciPy**
科学技術計算をサポート。
scikit-learnの内部で多用されてる。
統計処理とか、線形代数とか、フーリエ変換とか。すごい。
